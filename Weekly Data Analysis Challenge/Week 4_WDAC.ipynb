```python
# import libraries

import pandas as pd
import numpy as np
```

```python
# read in data:
df = pd.read_csv('https://buff.ly/3EtASMF')
```

```python
# view head of data
df.head()
```

```python
# view info
df.info()
```

### 1. See the first 10 entries.

```python
df.head(10)
```

### 2. Delete the columns 'Unnamed: 0' and 'Id'.

```python
df.drop(columns=['Unnamed: 0', 'Id'], inplace=True)
df
```

### 3. Group the dataset by name, assign to a variable called names, and sort the dataset by highest to lowest count.

```python
names = df.groupby('Name')['Count']\
          .sum()\
          .sort_values(ascending=False)\
          .reset_index()
names
```

### 4. How many different names exist in the dataset?

```python
df['Name'].nunique()
```

**There are 17,632 different names in the dataset.**

### 5. What is the name with most occurrences?

```python
names.head(1)
```

**Jacob is the name with the most occurrences.**

### 6. What is the standard deviation of count of names?

```python
names['Count'].std()
```

### 7. Get a summary of the dataset with the mean, min, max, std and quartiles.
