```python
# import libraries

import pandas as pd
import numpy as np
```

```python
# read in data:
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data' 
df = pd.read_csv(url)
```

```python
# view head of data
df.head(5)
```

```python
# view info
df.info()
```

### 1.    Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns

```python
df.drop(df.columns[[0, 3, 6, 8, 10, 12, 13]],axis = 1, inplace=True)
```

```python
df.head()
```

### 2.    Assign the columns as below:
- alcohol
- malic_acid
- alcalinity_of_ash
- magnesium
- flavanoids
- proanthocyanins
- hue

```python
df.columns=['alcohol', 'malic_acid', 'alcalinity_of_ash', 'magnesium', 'flavanoids', 'proanthocyanins', 'hue']
```

```python
df.head()
```

### 3.    Set the values of the first 3 rows in the alcohol column as NaN

```python
df.iloc[:3, 0] = np.nan
```

```python
df.head()
```

### 4.    Now set the value of the rows 3 and 4 of the magnesium column as NaN

```python
df.iloc[2:4, 3] = np.nan
```

```python
df.head(10)
```

### 5.    Fill in the null values (NaN) with the number 10 in the alcohol column and 100 in magnesium column

```python
df.iloc[:3, 0] = 10
df.iloc[2:4, 3] = 100
df.head()
```

### 6.    Count the number of missing values in the entire dataset

```python
df.isnull().sum()
```
